2.Create and apply fuction to read data by splitting by":::"

```python
def load_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = f.readlines()
    data = [line.strip().split(':::') for line in data]
    return data
```

```python
train_data = load_data(r"C:\Users\Vaibhav\Downloads\archive (1)\Genre Classification Dataset\train_data.txt") 
train_df = pd.DataFrame(train_data,columns=['ID','Title','Genre','Description'])

test_data = load_data(r"C:\Users\Vaibhav\Downloads\archive (1)\Genre Classification Dataset\test_data.txt") 
test_df = pd.DataFrame(test_data,columns=['ID','Title','Description'])

test_solution = load_data(r"C:\Users\Vaibhav\Downloads\archive (1)\Genre Classification Dataset\test_data_solution.txt") 
test_soultion_df= pd.DataFrame(test_solution,columns=['ID','Title','Genre','Description'])
```

3.Feature extraction : TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=10000)

X_train_tfidf = vectorizer.fit_transform(train_df["Description"])
X_test_tfidf = vectorizer.transform(test_df["Description"])

print(f"Training data shape:{X_train_tfidf.shape}")
print(f"Test data shape:{X_test_tfidf.shape}")
```

4.Encoding the Target labels

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_df['Genre'])
print(f"Unique genres in the training data: {label_encoder.classes_}")
```

5.Modle Buliding -Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression (max_iter=200)
lr_model.fit (X_train_tfidf, y_train)

y_pred = lr_model.predict(X_test_tfidf) 
predicted_genres = label_encoder.inverse_transform (y_pred )
test_df['Predicted_Genre'] = predicted_genres
test_df[['Title', 'Predicted_Genre']]
```

6.Model Evaluation - Logistic Regression

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report

# Create a sample DataFrame
data = {'Genre': ['Drama', 'Comedy', 'Action'],
        'predicted_genres': ['Drama', 'Action', 'Action']}
merged_df = pd.DataFrame(data)

# Calculate accuracy
accuracy = accuracy_score(merged_df['Genre'], merged_df['predicted_genres'])
print(f"accuracy: {accuracy:.4f}")

# Print classification report
print("\nClassification Report:")
print(classification_report(merged_df['Genre'], merged_df['predicted_genres']))
```

10.Modle Evaluation - SVM (Support Vector Manchine)

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# Sample data
data = {'text': ['The movie was great', 'I hated the film', 'It was a fantastic watch'],
        'genre': ['Positive', 'Negative', 'Positive']}
df = pd.DataFrame(data)

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(df['text'])

# Encode the labels
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(df['genre'])

# Define and train the SVM model
svm_model = SVC(kernel='linear')
svm_model.fit(X_train_tfidf, y_train)

# Sample test data
test_data = {'text': ['The film was amazing', 'I did not like the movie'],
             'ID': [1, 2]}
test_df = pd.DataFrame(test_data)

# Vectorize the test data
X_test_tfidf = vectorizer.transform(test_df['text'])

# Predict and inverse transform
y_pred_svm = svm_model.predict(X_test_tfidf)
predicted_genres_svm = label_encoder.inverse_transform(y_pred_svm)
test_df['Predicted_Genre_SVM'] = predicted_genres_svm

# Sample solution data for merging
test_solution_df = pd.DataFrame({'ID': [1, 2],
                                 'Genre': ['Positive', 'Negative']})

# Convert ID columns to the same type
test_solution_df['ID'] = test_solution_df['ID'].astype(str)
test_df['ID'] = test_df['ID'].astype(str)

# Merge DataFrames
merged_df_svm = pd.merge(test_solution_df, test_df[['ID', 'Predicted_Genre_SVM']], on='ID')

# Calculate accuracy
accuracy_svm = accuracy_score(merged_df_svm['Genre'], merged_df_svm['Predicted_Genre_SVM'])
print(f"SVM Accuracy: {accuracy_svm:.4f}")

# Print classification report using labels and target_names
print("\nSVM Classification Report:")
print(classification_report(merged_df_svm['Genre'], merged_df_svm['Predicted_Genre_SVM'], labels=label_encoder.classes_, target_names=label_encoder.classes_))
```
